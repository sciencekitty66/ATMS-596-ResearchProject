Unfortunately, FRANA data is not publicly available right now. FRANA is scheduled to transition into NWS operations later this year. When that happens the data will become public and archived on AWS. 

In the meantime, the best sources for verification that I'm aware of are point observations from ASOS and local storm reports (LSRs). Iowa State has a great website and API for accessing those archives.

I'm not sure what your aim is for this study, but if you want to create an "off-brand pseudo FRANA" - you can access the MRMS QPE in AWS here: https://noaa-mrms-pds.s3.amazonaws.com/index.html
You can also download the HRRR analyses from AWS as well, run the same ptype algorithm on it that you used for your WRF run, pair this with the MRMS data, and then run FRAM again on this data to get something similar to FRANA. That's a lot of work, but that would be the best way to try and recreate something similar to FRANA.

Hopefully this is helpful. And again, sorry the data is not public yet.
